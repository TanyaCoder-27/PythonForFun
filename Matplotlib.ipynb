```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {'Brand': ['NOISE', 'Fire-Boltt', 'boAt', 'pebble'],
        'PriceinRs': [2200, 1600, 1900, 2400]}
df = pd.DataFrame(data)

# Scatter plot
pp.scatter(df['Brand'], df['PriceinRs'])

# Add labels and title
pp.xlabel('Brand')
pp.ylabel('Price in Rs')
pp.title('Brand vs Price Scatter Plot')

# Show the plot
pp.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {'Brand': ['NOISE', 'Fire-Boltt', 'boAt', 'pebble'],
        'PriceinRs': [2200, 1600, 1900, 2400]}
df = pd.DataFrame(data)

# Define colors
colors = ['r', 'b', 'y', 'g']  # red, blue, yellow, green

# Create a pie chart
plt.pie(df['PriceinRs'], labels=df['Brand'], autopct='%1.1f%%',
        startangle=90, colors=colors)

# Equal aspect ratio ensures the pie chart is circular.
plt.axis('equal')

plt.title('Brand vs. Price Pie Chart')
plt.show()
```
